### Filling up the list with the relevant CODEM values and the DALYs from the central funcs

```python
import pandas as pd
causes_criteria = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/CandidateCauses_Criteria.xlsx')
```

```python
causes_criteria.head()
```

```python
causes_criteria = causes_criteria.dropna(subset=['GBD cause_id'])
```

```python
causes_criteria.columns
```

```python
causes_criteria.loc[:,'GBD cause_id'] = causes_criteria.loc[:,'GBD cause_id'].astype('int64')
```

```python
causes_criteria.shape
```

```python
causes_criteria.head()
```

## List of all cause and cause IDs from the source

```python
all_cause_data = pd.read_csv('/ihme/homes/arjuns13/notebooks/Documents/Data/all_cause_data.csv')
```

```python
all_cause_data = all_cause_data.drop(columns = ['acause', 'cause_description'])
```

```python
all_cause_data.head()
```

## COD list

```python
COD_cause_list = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/COD_methods_appendix_9_HAQ_Only.xlsx',\
                               header=2, na_values='--')
```

```python
COD_cause_list = COD_cause_list.dropna()
```

```python
COD_cause_list = COD_cause_list.drop(columns = ['Direction', 'Level', 'Age Start', 'Age End'])
```

```python
COD_cause_list.head()
```

```python
COD_cause_list.shape
```

## COD merged with All Cause list to get the cause_id's

```python
COD_cause_list_with_ids = COD_cause_list.merge(all_cause_data, left_on = 'Cause', right_on = 'cause_name')
```

```python
all_cause_data[all_cause_data['cause_name'] == 'Ischemic heart disease']
```

```python
COD_cause_list_with_ids = COD_cause_list_with_ids.drop(columns = ['cause_name'])
COD_cause_list_with_ids.head()
```

```python
COD_cause_list_with_ids.shape
```

```python
COD_cause_list_male = COD_cause_list_with_ids[COD_cause_list_with_ids['Sex'] == 'Male']
COD_cause_list_male.head()
```

```python
COD_cause_list_female = COD_cause_list_with_ids[COD_cause_list_with_ids['Sex'] == 'Female']
COD_cause_list_female.head()
```

```python
COD_cause_list_all = COD_cause_list_male.merge(COD_cause_list_female, on = ['cause_id', 'Cause', 'Model Type'])
COD_cause_list_all['Avg Num of Draws'] = (COD_cause_list_all['Number of Draws_x']\
                                              + COD_cause_list_all['Number of Draws_y'])/2
COD_cause_list_all = COD_cause_list_all.drop(columns = ['Sex_x', 'Sex_y', 'Number of Draws_x', 'Number of Draws_y' ])
COD_cause_list_all.head()
```

```python
COD_cause_list_all.shape
```

```python
#COD_cause_list_all = COD_cause_list_all.drop_duplicates(subset = 'cause_id')
```

```python
COD_cause_list_all.head()
```

## COD Data Rich model

```python
COD_cause_list_dataRich = COD_cause_list_all[COD_cause_list_all['Model Type'] == 'Data Rich']
COD_cause_list_dataRich.head()
```

```python
COD_cause_list_dataRich.shape
```

```python
COD_cause_list_dataRich = COD_cause_list_dataRich.drop_duplicates(subset = 'cause_id')
```

```python
COD_cause_list_dataRich[COD_cause_list_dataRich['Cause'] == 'Ischemic heart disease']
```

```python
COD_cause_list_dataRich.shape
```

```python
causes_criteria_final_sub = pd.merge(causes_criteria, COD_cause_list_dataRich,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final_sub.loc[:,'Number of GBD round 5data rich CODEM models in which HAQ index was included'] = \
causes_criteria_final_sub.loc[:,'Avg Num of Draws']
causes_criteria_final_sub = causes_criteria_final_sub.drop(columns = ['Cause', 'Model Type', 'cause_id', 'Avg Num of Draws'])
causes_criteria_final_sub.head()
```

## COD Global model

```python
COD_cause_list_global = COD_cause_list_all[COD_cause_list_all['Model Type'] == 'Global']
COD_cause_list_global.head(20)
```

```python
COD_cause_list_global.shape
```

```python
COD_cause_list_global = COD_cause_list_global.drop_duplicates(subset = 'cause_id')
```

```python
COD_cause_list_global.shape
```

```python
causes_criteria_final = pd.merge(causes_criteria_final_sub, COD_cause_list_global,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final.loc[:,'Number of GBD round 5 global CODEM models in which HAQ index was included'] = \
causes_criteria_final.loc[:,'Avg Num of Draws']
causes_criteria_final = causes_criteria_final.drop(columns = ['Cause', 'Model Type', 'cause_id', 'Avg Num of Draws'])
causes_criteria_final.head()
```

```python
causes_criteria_final.to_csv('/ihme/homes/arjuns13/notebooks/Documents/Data/CandidateCauses_Criteria_withCODEM.csv')
```

```python
cause_list = causes_criteria.loc[:,'GBD cause_id']
```

```python
cause_list = list(cause_list)
```

## And now onto the DALYs 

### First DALY rate in the US

```python
from db_queries import get_outputs as go 
dalys_per_cause_rate = go("cause", cause_id=cause_list, metric_id=3,\
                     measure_id=2, gbd_round_id=5,year_id=[2016], location_id=[102],\
                    age_group_id = 22)
dalys_per_cause_rate.head()
```

```python
dalys_per_cause_rate = dalys_per_cause_rate[['cause_id', 'val']]
dalys_per_cause_rate.head()
```

```python
causes_criteria_final1 = pd.merge(causes_criteria_final, dalys_per_cause_rate,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final1.loc[:,'DALYs per person in the United States, GBD round 5, year = 2016'] = \
causes_criteria_final1.loc[:,'val']
causes_criteria_final1 = causes_criteria_final1.drop(columns = ['cause_id', 'val'])
causes_criteria_final1.head()
```

### Next DALY total in the US

```python
dalys_per_cause_total = go("cause", cause_id=cause_list, metric_id=1,\
                     measure_id=2, gbd_round_id=5,year_id=[2016], location_id=[102],\
                    age_group_id = 22)
dalys_per_cause_total = dalys_per_cause_total[['cause_id', 'val']]

causes_criteria_final2 = pd.merge(causes_criteria_final1, dalys_per_cause_total,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final2.loc[:,'Total DALYs in the United States, GBD round 5, year = 2016'] = \
causes_criteria_final2.loc[:,'val']
causes_criteria_final2 = causes_criteria_final2.drop(columns = ['cause_id', 'val'])
causes_criteria_final2.head()
```

### Next DALY rate globally

```python
dalys_per_cause_rate_global = go("cause", cause_id=cause_list, metric_id=3,\
                     measure_id=2, gbd_round_id=5,year_id=[2016], location_id=[1],\
                    age_group_id = 22)
dalys_per_cause_rate_global = dalys_per_cause_rate_global[['cause_id', 'val']]
#dalys_per_cause_rate_global.head()
causes_criteria_final3 = pd.merge(causes_criteria_final2, dalys_per_cause_rate_global,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final3.loc[:,'DALYs per person globally, GBD round 5, year = 2016'] = \
causes_criteria_final3.loc[:,'val']
causes_criteria_final3 = causes_criteria_final3.drop(columns = ['cause_id', 'val'])
causes_criteria_final3.head()
```

### DALYs total globally

```python
dalys_per_cause_total_global = go("cause", cause_id=cause_list, metric_id=1,\
                     measure_id=2, gbd_round_id=5,year_id=[2016], location_id=[1],\
                    age_group_id = 22)
dalys_per_cause_total_global = dalys_per_cause_total_global[['cause_id', 'val']]
#dalys_per_cause_rate_global.head()
causes_criteria_final4 = pd.merge(causes_criteria_final3, dalys_per_cause_total_global,\
                                 left_on = 'GBD cause_id', right_on = 'cause_id', how = 'left')
causes_criteria_final4.loc[:,'Total DALYs globally, GBD round 5, year = 2016'] = \
causes_criteria_final4.loc[:,'val']
causes_criteria_final4 = causes_criteria_final4.drop(columns = ['cause_id', 'val'])
causes_criteria_final4.head()
```
