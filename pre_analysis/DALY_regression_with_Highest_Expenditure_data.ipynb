## Load and analyze the data for the causes with highest expenditure in the US (Joe's list)

```python
import pandas as pd
dex_cause_list = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/DEX_Cause list.xlsx')
```

```python
dex_cause_list.head()
```

```python
dex_cause_list = dex_cause_list.drop(columns = ['Aggregated health conditions']).rename(index=str, \
                                                                                        columns={"Unnamed: 0": "Rank",\
                                                                                                 "Most granular causes for disease expenditure research": "Cause"})
```

```python
dex_cause_list.shape
```

```python
dex_cause_list.head()
```

```python
COD_cause_list = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/COD_methods_appendix_9_HAQ_Only.xlsx',\
                               header=2, na_values='--')
```

```python
COD_cause_list = COD_cause_list.dropna()
```

```python
#COD_cause_list = COD_cause_list[COD_cause_list['Direction'] == -1]
```

```python
COD_cause_list = COD_cause_list.drop(columns = ['Direction', 'Level'])
```

```python
COD_cause_list.info()
```

```python
COD_cause_list.head()
```

```python
COD_cause_list.shape
```

```python
COD_cause_list_dataRich = COD_cause_list[COD_cause_list['Model Type'] == 'Data Rich']
COD_cause_list_dataRich.head()
```

```python
COD_cause_list_dataRich.shape
```

```python
COD_cause_list_Global = COD_cause_list[COD_cause_list['Model Type'] == 'Global']
COD_cause_list_Global.head()
```

```python
COD_cause_list_Global.shape
```

```python
COD_cause_list_all = COD_cause_list_dataRich.merge(COD_cause_list_Global, on = ['Cause', 'Sex', 'Age Start',\
                                                                                'Age End'])
```

```python
COD_cause_list_all.head()
```

```python
COD_cause_list_all.shape
```

```python
COD_cause_list_all['Avg Num of Draws'] = (COD_cause_list_all['Number of Draws_x'] + COD_cause_list_all['Number of Draws_y'])/2
```

```python
COD_cause_list_all = COD_cause_list_all.drop(columns = ['Age Start', 'Age End', 'Model Type_x', 'Number of Draws_x',\
                                              'Model Type_y', 'Number of Draws_y'])
```

```python
COD_cause_list_all = COD_cause_list_all.sort_values(by = ['Avg Num of Draws'], ascending=False)
```

```python
COD_cause_list_all.head()
```

```python
COD_cause_list_male = COD_cause_list_all[COD_cause_list_all['Sex'] == 'Male']
COD_cause_list_male.head()
```

```python
COD_cause_list_male.shape
```

```python
COD_cause_list_female = COD_cause_list_all[COD_cause_list_all['Sex'] == 'Female']
COD_cause_list_female.head()
```

```python
COD_cause_list_female.shape
```

```python
COD_cause_list_all_all = COD_cause_list_male.merge(COD_cause_list_female, on = ['Cause'])
```

```python
COD_cause_list_all_all.head()
```

```python
COD_cause_list_all_all.shape
```

```python
COD_cause_list_all_all['Avg Num of Draws'] = (COD_cause_list_all_all['Avg Num of Draws_x']\
                                              + COD_cause_list_all_all['Avg Num of Draws_y'])/2
```

```python
COD_cause_list_all_all = COD_cause_list_all_all.drop(columns = ['Avg Num of Draws_x',\
                                              'Sex_x', 'Sex_y', 'Avg Num of Draws_y'])
```

```python
COD_cause_list_all_all = COD_cause_list_all_all.sort_values(by = ['Avg Num of Draws'], ascending=False)
```

```python
COD_cause_list_all_all.head()
```

## Now that we have all the relevant causes, let's find their corresponding cause-IDs

```python
all_cause_data = pd.read_csv('/ihme/homes/arjuns13/notebooks/Documents/Data/all_cause_data.csv')
```

```python
all_cause_data.head(20)
```

```python
COD_cause_list_with_ids = COD_cause_list_all_all.merge(all_cause_data, left_on = 'Cause', right_on = 'cause_name')
```

```python
COD_cause_list_with_ids.head()
```

```python
dex_cause_list = pd.read_csv('/ihme/homes/arjuns13/notebooks/Documents/Data/dex_to_gbd_cause_match.csv')
```

```python
final_cause_list = final_cause_list.drop(columns=['dex_cause','gbd_cause','cause_name_x','Avg Num of Draws','cause_name_y'])
```
