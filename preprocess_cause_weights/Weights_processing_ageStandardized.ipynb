# This notebook takes as input the excel worksheet which has all the new cause weights and pre-processes it in a way that can then be directly consumed by the HAQ-index-generation code. This notebook deals with the age standardized version.

```python
import pandas as pd
pre_weights = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/HAQ causes weights by age category_GBD2017_sinarj.xlsx',\
                           skiprows = 4)
```

```python
pre_weights.head(20)
```

```python
pre_weights.columns
```

```python
pre_weights_processed_ageStd = pre_weights[['GBD age category ', 'Cause ID', 'Cause weight']]
pre_weights_processed_ageStd.head()
```

```python
pre_weights_processed_ageStd = pre_weights_processed_ageStd.drop(pre_weights_processed_ageStd.index[-1])
```

```python
pre_weights_processed_ageStd['Cause ID'] = pre_weights_processed_ageStd['Cause ID'].astype(int)
pre_weights_processed_ageStd.head()
```

```python
sum(pre_weights_processed_ageStd['Cause weight'])
```

```python
pre_weights_processed_ageStd = pre_weights_processed_ageStd.rename(index=str, columns={"GBD age category ": "cause_name",\
                                                        "Cause ID": "cause_id", "Cause weight": "cause_weight" }) 
```

```python
pre_weights_processed_ageStd.to_csv('/share/scratch/projects/hssa/haq/HAQ_2017/haq_US/weights/cause_weights_ageStd_2017.csv',\
                            index = False)
```
