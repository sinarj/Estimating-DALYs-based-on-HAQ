# This notebook takes as input the excel worksheet which has all the new cause weights and pre-processes it in a way that can then be directly consumed by the notebook "Weights_processing_ageSpecific" which will then prepare it further for the HAQ-index-generation code. This notebook deals with the age specific version. 

```python
import pandas as pd
pre_weights = pd.read_excel('/ihme/homes/arjuns13/notebooks/Documents/Data/HAQ causes weights by age category_GBD2017_sinarj.xlsx',\
                           skiprows = 4)
```

```python
pre_weights.head(20)
```

```python
pre_weights_processed = pre_weights[['GBD age category ', 'Cause ID', 'Rescaled Weights  0-6 days', ]]
```

```python
pre_weights_processed = pre_weights.iloc[:, [1,2,8,11,14,17,20,23,26,29,32,35,38,41,44,47,50,53,56,59] ]
```

```python
pre_weights_processed
```

```python
col_names = ['cause_name', 'cause_id']
for i in range(2,20):
    col_names.append(i)
col_names
```

```python
pre_weights_processed.columns = col_names
```

```python
pre_weights_processed = pre_weights_processed.drop(pre_weights_processed.index[-1])
```

```python
pre_weights_processed.cause_id = pre_weights_processed.cause_id.astype(int)
```
