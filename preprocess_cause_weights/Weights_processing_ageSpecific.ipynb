## Setting cause weight = 0 for Diptheria, Maternal disorders and CHA and rescaling

```python
weights = pd.read_csv(data_dr+'process/cause_weights_by_age.csv')
```

```python
## new cause weights for age groups 15,16,17,18 will be same as that of age group 19
tmp = weights[weights['age_group_id']==19]
```

```python
weights = weights.merge(tmp[['cause_id','cause_weight']],on='cause_id')
```

```python
weights.loc[weights['age_group_id']>14,'cause_weight_x'] = weights.loc[weights['age_group_id']>14,'cause_weight_y']
weights.drop(['cause_weight_y'],axis=1,inplace=True)
weights = weights.rename(columns={'cause_weight_x':'cause_weight'})
```

```python
weights.to_csv(data_dr+'csvs/cause_weights_by_age_v2.csv',index=False)
```
